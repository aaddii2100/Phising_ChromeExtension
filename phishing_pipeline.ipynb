```python
def predict_new_email(email_text):
    # The pipeline automatically cleans, vectorizes, and predicts
    prediction = pipeline.predict([email_text])[0]
    probability = pipeline.predict_proba([email_text])[0][1]  # probability for phishing class
    return prediction, probability

# Example new email:
new_email = "Alert: Your bank account has been suspended. Click here to resolve: http://suspicious-link.com"
pred, score = predict_new_email(new_email)
if score >= 0.5:
    print(f"Phishing alert! Score: {score:.2f}. The email is predicted as phishing.")
else:
    print(f"Email appears legitimate. Score: {score:.2f}.")
```
