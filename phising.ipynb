```python
# preprocessing the text col

import re
import pandas as pd
from nltk.corpus import stopwords
import nltk

# Download stopwords if not already available
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))

# Function to clean text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"http\S+|www\S+", "", text)  # Remove URLs
    text = re.sub(r"[^\w\s]", "", text)  # Remove special characters
    text = re.sub(r"\s+", " ", text).strip()  # Remove extra spaces
    text = " ".join(word for word in text.split() if word not in stop_words)  # Remove stopwords
    return text

# Apply cleaning
df["clean_text"] = df["text"].apply(clean_text)
```

```python
df
```

```python
# dropping the text col
df.drop(columns=['text'],inplace=True)
df
```

```python
# rearranging
df =df[['clean_text','label']]
df
```

```python
# converting texts into numbers and also train_test_split

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df["clean_text"], df["label"], test_size=0.2, random_state=42)

# Convert text into TF-IDF vectors
tfidf = TfidfVectorizer(max_features=5000)  # Limit to 5000 words for efficiency
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
X_train_tfidf
```

```python
# logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Train model
log_reg = LogisticRegression()
log_reg.fit(X_train_tfidf, y_train)

# Predict on test data
y_pred_log_reg = log_reg.predict(X_test_tfidf)

# Evaluate model
print("Logistic Regresssion Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print(classification_report(y_test, y_pred_log_reg))
```

```python
#  random forest

from sklearn.ensemble import RandomForestClassifier

# Train Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_tfidf, y_train)

# Predict
y_pred_rf = rf.predict(X_test_tfidf)

# Evaluate
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
# svm

from sklearn.svm import SVC

# Train SVM
svm = SVC(kernel="linear", probability=True)
svm.fit(X_train_tfidf, y_train)

# Predict
y_pred_svm = svm.predict(X_test_tfidf)

# Evaluate
print("SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python
# xgboost

from sklearn.ensemble import GradientBoostingClassifier

# Train Gradient Boosting
gb = GradientBoostingClassifier(n_estimators=100)
gb.fit(X_train_tfidf, y_train)

# Predict
y_pred_gb = gb.predict(X_test_tfidf)

# Evaluate
print("✅ Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gb))
print(classification_report(y_test, y_pred_gb))
```

```python
# naive bayes

from sklearn.naive_bayes import MultinomialNB

# Train Naïve Bayes
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)

# Predict
y_pred_nb = nb.predict(X_test_tfidf)

# Evaluate
print("Naïve Bayes Accuracy:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
```

```python
from sklearn.metrics import accuracy_score

# Get training predictions
y_train_pred_log_reg = log_reg.predict(X_train_tfidf)
y_train_pred_rf = rf.predict(X_train_tfidf)
y_train_pred_svm = svm.predict(X_train_tfidf)
y_train_pred_gb = gb.predict(X_train_tfidf)
y_train_pred_nb = nb.predict(X_train_tfidf)

# Print accuracy scores
print("🔹 Logistic Regression: Train =", accuracy_score(y_train, y_train_pred_log_reg), "| Test =", accuracy_score(y_test, y_pred_log_reg))
print("🔹 Random Forest: Train =", accuracy_score(y_train, y_train_pred_rf), "| Test =", accuracy_score(y_test, y_pred_rf))
print("🔹 SVM: Train =", accuracy_score(y_train, y_train_pred_svm), "| Test =", accuracy_score(y_test, y_pred_svm))
print("🔹 Gradient Boosting: Train =", accuracy_score(y_train, y_train_pred_gb), "| Test =", accuracy_score(y_test, y_pred_gb))
print("🔹 Naïve Bayes: Train =", accuracy_score(y_train, y_train_pred_nb), "| Test =", accuracy_score(y_test, y_pred_nb))
```

```python
# Checking accuracy score of all the models

from sklearn.model_selection import cross_val_score

# Define models in a dictionary
models = {
    "Logistic Regression": log_reg,
    "Random Forest": rf,
    "SVM": svm,
    "Gradient Boosting": gb,
    "Naïve Bayes": nb
}

# Perform cross-validation for each model
for name, model in models.items():
    scores = cross_val_score(model, X_train_tfidf, y_train, cv=10, scoring='accuracy')

    print(f"🔹 {name} → Cross-Validation Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Define parameter grid for SVM
svm_param_grid = {
    "C": [0.1, 1, 10, 100],  # Regularization strength
    "kernel": ["linear", "rbf"],  # Kernel type
    "gamma": ["scale", "auto"]  # Kernel coefficient
}

# Create SVM model with probability estimation enabled
svm_model = SVC(probability=True)  # 🔹 Enable probability estimation

# Perform Grid Search
svm_grid_search = GridSearchCV(svm_model, svm_param_grid, cv=5, scoring="accuracy", n_jobs=-1)
svm_grid_search.fit(X_train_tfidf, y_train)

# Best parameters and accuracy
print("🔹 Best SVM Parameters:", svm_grid_search.best_params_)
print("✅ Best SVM Accuracy:", svm_grid_search.best_score_)

# 🔹 Save the best SVM model
best_svm_model = svm_grid_search.best_estimator_
```

```python
# Hyperparameter tuning with naive bayes


from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV

# Define parameter grid
nb_param_grid = {
    "alpha": [0.1, 0.5, 1, 5, 10]  # Smoothing values
}

# Create Naïve Bayes model
nb_model = MultinomialNB()

# Perform Grid Search
nb_grid_search = GridSearchCV(nb_model, nb_param_grid, cv=5, scoring="accuracy", n_jobs=-1)
nb_grid_search.fit(X_train_tfidf, y_train)

# Best parameters and accuracy
print("🔹 Best Naïve Bayes Parameters:", nb_grid_search.best_params_)
print("✅ Best Naïve Bayes Accuracy:", nb_grid_search.best_score_)
```
